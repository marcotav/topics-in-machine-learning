## Cleaning up data
[[go back to the top]](#Table-of-contents)

```python
shopping.head()
loja.head()
varej.head()
```

We will need the following function to remove accents:

```python
import unidecode

def un_st(x):
    return unidecode.unidecode(x)
```

```python
shopping, varej = shopping.iloc[:,1:-1], varej.iloc[:,1:7]
shopping.head()
loja.head()
varej.head()
```

```python
for el in [shopping, loja, varej]:
    el.dropna(inplace=True)
    el.columns = [un_st(x.lower().replace(' ','_').replace('(','').replace(')',''))
                        for x in el.columns]

shopping.head()
loja.head()
varej.head()
```

```python
cols_to_dummy = ['piso', 'corredor', 
                 'tipo_de_loja', 'loja_esquerda', 
                 'loja_direita', 'loja_em_frente', 'esquina',
                 'segmento', 'tipo_de_segmento']

for col in cols_to_dummy:
    aux.s_to_df(loja[col].value_counts(), col,'frequency').head()
```

```python
cols_to_dummy.remove('piso')
```

```python
print(cols_to_dummy + ['piso', 'ocupante'])
```

```python
dummies = pd.get_dummies(loja[cols_to_dummy], drop_first=True)  
```

```python
dummies.columns = [un_st(x.lower().replace(' ','_')) for x in dummies.columns]
```

```python
loja = loja.drop(cols_to_dummy + ['piso'], axis=1)
loja.head()
```

```python
loja = pd.concat([loja, dummies], axis=1)
# loja.set_index('loja', inplace=True)
loja.head()
```
